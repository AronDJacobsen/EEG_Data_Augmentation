```python
import os
import numpy as np
import pandas as pd
import mne
import matplotlib.pyplot as plt
from PIL import Image
from random import uniform
from IPython.utils import io
import random
```

```python
path_xlsx_file = 'dataEEG/'
file_name_xlsx = 'MGH_File_Annotations.xlsx'
annotation = pd.read_excel(path_xlsx_file+file_name_xlsx, sheet_name= 3)
annotation.head()
```

```python
# Only need to run this cell the first time you use the script

annotation_path = annotation['Recording']
annotation_Quality = annotation['Quality Of Eeg']

file_names = []
quality = []

for i in range(len(annotation_path)):
    if type(annotation_Quality[i]) == int:
        file_names.append(annotation_path[i].split('/')[-1])
        quality.append(int(int(annotation_Quality[i]) > 7))
```

```python
path = 'dataEEG/'
fileName_quality = []

idx = None

for folder, subfolder, filenames in os.walk(path):
    for edf_file in filenames:
        
        idx = None
        
        #search for filename index
        for i, file_name in enumerate(file_names):
            if file_name == edf_file:
                idx = i
                break
        
        # create list of tuples contaning the patg to the fil and gender
        if idx != None and quality[idx]==1:
            fileName_quality.append((folder+'/'+edf_file, quality[idx]))
        else:
            continue
```

```python
###############################################################################################################################
###############################################################################################################################
###############################################################################################################################

################################# ONLY RUN THIS CELL ONCE OTHERWISE YOU DOUBLICATE THE IMAGES #################################

###############################################################################################################################
###############################################################################################################################
###############################################################################################################################

#breaks = 0
window_size = 30
count = 0
channel = 3

# Creating data
for path, q in fileName_quality:
    with io.capture_output() as captured:
        raw  = mne.io.read_raw_edf(path);
        freq = raw.info['sfreq'];
        num_samples = raw.n_times
        ch_names = raw.ch_names
        ch_types = ('eeg,'*14).split(',')[0:-1]
        info = mne.create_info(ch_names=ch_names, sfreq=freq, ch_types=ch_types)
    lower_bound = list(range(0, int(num_samples/freq)-window_size, window_size-int(window_size/2)))
    upper_bound = list(range(window_size, int(num_samples/freq), window_size-int(window_size/2)))
    
    
    for low, up in zip(lower_bound,upper_bound):
        # Copy the data so we stil have the full data file after crop
        with io.capture_output() as captured:
            raw_subset_norm = raw.copy();
            raw_subset_ra = raw.copy();
            raw_subset_wn = raw.copy();
            raw_subset_fo = raw.copy();
            
        # crop the copy
        with io.capture_output() as captured:
            raw_subset_norm = raw_subset_norm.crop(tmin = low, tmax = up).load_data();
            raw_subset_ra = raw_subset_ra.crop(tmin = low, tmax = up).load_data();
            raw_subset_wn = raw_subset_wn.crop(tmin = low, tmax = up).load_data();
            raw_subset_fo = raw_subset_fo.crop(tmin = low, tmax = up).load_data();
        
        #Reduce amplitude
        data_ra = raw_subset_ra.get_data()
        data_ra = data_ra/1000 

        
        #add white noise
        data_wn = raw_subset_wn.get_data()
        data_wn = data_wn+5*np.random.randn(data_wn.shape[0], data_wn.shape[1])
        
        #create fall out
        data_fo = raw_subset_fo.get_data()
        
        fall_of_length= random.randint(500,data_fo.shape[1])
        start_fall_of = random.randint(1,(data_fo.shape[1]-fall_of_length-1))
        l1 = ('1,'*start_fall_of).split(',')[0:-1]
        l2 = ('0.0000000000000000001,'*fall_of_length).split(',')[0:-1]
        l3 = ('1,'*(data_fo.shape[1]-start_fall_of-fall_of_length)).split(',')[0:-1]
        fall_out = np.array(l1+l2+l3, dtype='float')
        
        data_fo = data_fo*fall_out

        
        with io.capture_output() as captured:
            raw_subset_fo = mne.io.RawArray(data_fo, info)
            raw_subset_wn = mne.io.RawArray(data_wn, info)
            raw_subset_ra = mne.io.RawArray(data_ra, info)
        
        # filter the data
        with io.capture_output() as captured:
            raw_subset_norm.filter(l_freq=0.4, h_freq=40);
            raw_subset_ra.filter(l_freq=0.4, h_freq=40)
            raw_subset_wn.filter(l_freq=0.4, h_freq=40)
            raw_subset_fo.filter(l_freq=0.4, h_freq=40)
            
        # extract the data values
        with io.capture_output() as captured:
            data_norm = raw_subset_norm.get_data();
            data_ra = raw_subset_ra.get_data()
            data_wn = raw_subset_wn.get_data()
            data_fo = raw_subset_fo.get_data()
            
        # Choose channel
        data_norm = data_norm[channel]
        data_ra = data_ra[channel]
        data_wn = data_wn[channel]
        data_fo = data_fo[channel]
        
        # create name
        Img_name = str(count)+'_'+str(channel)+'.jpg'
        
        #creat normal plot
        pxx, freqs, bins, im = plt.specgram(data_norm, Fs = freq, cmap = 'gray')
        pxx = np.log1p(pxx)
        plt.axis('off')
        
        # decide between test or train
        test_or_train = uniform(0, 1)
        
        # save normal data
        if test_or_train < 0.9:
            plt.savefig('D:/qd/Train/Good/'+Img_name, edgecolor = None, bbox_inches='tight',pad_inches = 0)
        else:
            plt.savefig('D:/qd/Test/Good/'+Img_name, edgecolor = None, bbox_inches='tight',pad_inches = 0)
        
        plt.close()
        
        # create white reduced amp plot
        pxx, freqs, bins, im = plt.specgram(data_ra, Fs = freq, cmap = 'gray')
        pxx = np.log1p(pxx)
        plt.axis('off')
        
        # decide between test or train
        test_or_train = uniform(0, 1)
        
        # save reduced amp plot
        if test_or_train < 0.9:
            plt.savefig('D:/qd/Train/Reduced_amp/'+Img_name, edgecolor = None, bbox_inches='tight',pad_inches = 0)
        else:
            plt.savefig('D:/qd/Test/Reduced_amp/'+Img_name, edgecolor = None, bbox_inches='tight',pad_inches = 0)
        
        plt.close()  
        
        # create white noise plot
        pxx, freqs, bins, im = plt.specgram(data_wn, Fs = freq, cmap = 'gray')
        pxx = np.log1p(pxx)
        plt.axis('off')
        
        # decide between test or train
        test_or_train = uniform(0, 1)
        
        # save white noise plot plot
        if test_or_train < 0.9:
            plt.savefig('D:/qd/Train/White_Noise/'+Img_name, edgecolor = None, bbox_inches='tight',pad_inches = 0)
        else:
            plt.savefig('D:/qd/Test/White_Noise/'+Img_name, edgecolor = None, bbox_inches='tight',pad_inches = 0)
        
        plt.close()         
        
        # create fall out plot
        pxx, freqs, bins, im = plt.specgram(data_fo, Fs = freq, cmap = 'gray')
        pxx = np.log1p(pxx)
        plt.axis('off')
        
        # decide between test or train
        test_or_train = uniform(0, 1)
        
        # save fall out plot
        if test_or_train < 0.9:
            plt.savefig('D:/qd/Train/Fall_Out/'+Img_name, edgecolor = None, bbox_inches='tight',pad_inches = 0)
        else:
            plt.savefig('D:/qd/Test/Fall_Out/'+Img_name, edgecolor = None, bbox_inches='tight',pad_inches = 0)
        
        plt.close()    
        count +=1
        
        del raw_subset_norm
        del raw_subset_fo
        del raw_subset_ra
        del raw_subset_wn
        del data_norm
        del data_fo
        del data_ra
        del data_wn
    del raw

            
```
